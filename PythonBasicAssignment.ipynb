```python
2. Describe the role of predefined keywords in python and provide examples of how they are used in a program.

Answer: Predefined keywords in Python are reserved words that have a specific meaning and purpose in the language.
        They are part of Python's syntax and cannot be used for variable names, function names, or any other identifier. 

=> Here are some key predefined keywords and their uses:

---------------------------------------------------------------------------------------------------
Example 1: if, elif, else: Used for conditional statements.
    
age = 20
if age < 18:
    print("Minor")
elif age == 18:
    print("Just an adult")
else:
    print("Adult")

-------------------------------------------------------------------------------------------------------
Example 2: for, while, break, continue: Used for loops and controlling loop execution.

=> for loop:
for i in range(5):
    if i == 3:
        continue
    print(i)

=> while loop:
n = 0
while n < 5:
    if n == 3:
        break
    print(n)
    n += 1

-----------------------------------------------------------------------------------------------------------
Example 3: def, return: Used to define functions and return values from them.

def greet(name):
    return f"Hello, {name}!"
print(greet("Alice"))

------------------------------------------------------------------------------------------------------------
Example 4: class, self: Used for defining classes and referring to instance variables.
    
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        return f"{self.name} says woof!"

my_dog = Dog("Buddy")
print(my_dog.bark())
import, from, as: Used for importing modules.
    import math
from math import sqrt as square_root

print(math.pi)
print(square_root(16))

-------------------------------------------------------------------------------------------------------------
Example 5: try, except, finally: Used for handling exceptions.

try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero!")
finally:
    print("This will always execute.")
```

```python
3. Compare and contrast mutable and immutable objects in python with examples.

Answer: 
---------------------------------------------------------------------------------------------------------------------------------    
Mutable objects can be changed after their creation. This means you can modify their content without creating a new object.
---------------------------------------------------------------------------------------------------------------------------------

Examples: Lists, dictionaries, sets.

Example:

# List (Mutable)
my_list = [1, 2, 3]
print("Original list:", my_list)

# Modifying the list
my_list.append(4)
print("Modified list:", my_list)

Output:
Original list: [1, 2, 3]
Modified list: [1, 2, 3, 4]
--------------------------------------------------------------------------------------------------------------------------------
Immutable objects cannot be changed once they are created. Any modification to an immutable object results in a new object.
--------------------------------------------------------------------------------------------------------------------------------

Examples: Tuples, strings, numbers (integers, floats).

Example:

# String (Immutable)
my_string = "Hello"
print("Original string:", my_string)

# Trying to modify the string
new_string = my_string + " World"
print("New string:", new_string)

Output:
Original string: Hello
New string: Hello World

----------------------
Key Differences:
----------------------
a) Changeability:
   Mutable: Elements can be added, removed, or altered.
   Immutable: Any operation that seems to modify the object actually creates a new object.

b) Performance:
   Mutable: More efficient for large data sets that require frequent updates since it doesn't need to create new objects.
   Immutable: Often faster for read-only operations since they can be shared safely among multiple threads without needing locks.
    
c) Usage:
   Mutable: Suitable for collections that need frequent modifications, like lists of user inputs or dynamic sets of data.
   Immutable: Ideal for constants, fixed data sets, or when you want to ensure data integrity throughout the program execution.
```

```python
5. Explain the concept of type casting in python with examples.

Answer: 
    Type casting in Python is the process of converting one data type to another. This is useful when you want to perform operations that 
    require a specific data type. 
    There are two types of type casting: implicit and explicit.
--------------------------------------------------------------------------------------------------------------------------
Implicit Type Casting:
This is done automatically by Python, where it converts a smaller data type to a larger data type to prevent data loss.
--------------------------------------------------------------------------------------------------------------------------

Example:

# Implicit type casting
num_int = 10   # Integer
num_float = 10.5  # Float

# Addition of integer and float
result = num_int + num_float
print("Data type of num_int:", type(num_int))
print("Data type of num_float:", type(num_float))
print("Result:", result)
print("Data type of result:", type(result))

Output:
Data type of num_int: <class 'int'>
Data type of num_float: <class 'float'>
Result: 20.5
Data type of result: <class 'float'>

In this example, Python implicitly converts the integer num_int to float before performing the addition.

----------------------------------------------------------------------------------------------------------------
Explicit Type Casting:
This is done manually by the programmer using predefined functions like int(), float(), str(), etc.
----------------------------------------------------------------------------------------------------------------

Example:

# Explicit type casting
num_str = "100"  # String

# Convert string to integer
num_int = int(num_str)
print("Data type of num_str:", type(num_str))
print("Data type of num_int after casting:", type(num_int))

# String to float
num_float = float(num_str)
print("Data type of num_float after casting:", type(num_float))

Output:
Data type of num_str: <class 'str'>
Data type of num_int after casting: <class 'int'>
Data type of num_float after casting: <class 'float'>

In this example, the string num_str is explicitly converted to an integer and a float.
```

```python
6. How do conditional statements work in python? Illustrate with examples.

Answer: Conditional statements in Python allow you to execute different blocks of code based on certain conditions. 
        The most common conditional statements are if, elif (short for else if), and else.

------------------
Basic Structure:
------------------

if statement: Executes a block of code if the condition is true.

elif statement: Checks another condition if the previous ones are false.

else statement: Executes a block of code if all previous conditions are false.

-------------------------------------
Here's a breakdown with examples:
-------------------------------------

Example 1: Simple if Statement

age = 18

if age >= 18:
    print("You are an adult.")

Output:
You are an adult.

--------------------------------------
Example 2: if-elif-else Statement
python
temperature = 25

if temperature > 30:
    print("It's a hot day.")
elif temperature >= 20:
    print("It's a nice day.")
else:
    print("It's a bit chilly.")

Output:
It's a nice day.

* Conditional statements help control the flow of our programs, making them more dynamic and responsive to different inputs.
```

```python
7. Describes the different types of Loops in python and their use cases with example.

Answer: Python, loops are used to repeatedly execute a block of code. 
        There are two main types of loops: for loops and while loops.
-------------
a) For Loop
-------------
A for loop is used to iterate over a sequence (like a list, tuple, dictionary, set, or string) and 
execute a block of code for each item in the sequence.

Use Cases: Iterating through items in a collection, performing operations on each item, generating sequences.

-------------
Example :
-------------
 
Iterating through a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

Output:
apple
banana
cherry

-----------------
b) While Loop
-----------------
A while loop continues to execute a block of code as long as a specified condition is true.

Use Cases: When the number of iterations is not known beforehand, processing input until a certain condition is met.

-----------
Example:
-----------

python
# Print numbers from 1 to 5
i = 1
while i <= 5:
    print(i)
    i += 1
Output:
1
2
3
4
5

----------------------------
Loop Control Statements:
----------------------------
break: Terminates the loop prematurely.
continue: Skips the rest of the code inside the loop for the current iteration and moves to the next iteration.

-----------------
Nested Loops:
----------------
Loops inside another loop. They are used for more complex iterations, such as iterating through a matrix or creating combinations.
Use Cases: Iterating through multidimensional arrays, creating patterns.
```
