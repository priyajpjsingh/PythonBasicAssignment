```python
Q1. Disscuss String slicing and provide examples.
Answer: String slicing is a powerful feature in Python that allows you to extract a portion of a string using a specific syntax. 
        
Basic Syntax:

The basic syntax for slicing a string is:
string[start:end:step]
start (optional) is the index where the slice starts (inclusive).
end (optional) is the index where the slice ends (exclusive).
step (optional) is the step size or stride.
                
Examples:
Simple Slicing Extract a substring from index 2 to 5.
-----------------------------------------------------
python
text = "Hello, World!"
slice = text[2:5]
print(slice)
Output:
llo

Omitting Start Index If you omit the start index, slicing starts from the beginning of the string.
---------------------------------------------------------------------------------------------------
python
slice = text[:5]
print(slice)
Output:
Hello

Omitting End Index If you omit the end index, slicing goes until the end of the string.
----------------------------------------------------------------------------------------
python
slice = text[7:]
print(slice)
Output:
World!

Using Negative Indexes Negative indexes count from the end of the string.
-------------------------------------------------------------------------
python
slice = text[-6:-1]
print(slice)
Output:
World

Step in Slicing The step parameter defines the increment between each index. For example, to get every second character:
------------------------------------------------------------------------------------------------------------------------
python
slice = text[::2]
print(slice)
Output:
Hlo ol!

Reversing a String Using a negative step, you can reverse a string.
-------------------------------------------------------------------
python
reversed_text = text[::-1]
print(reversed_text)
Output:
!dlroW ,olleH

Real-World Use Cases Extracting a Substring
---------------------------------------------

python
date = "2024-11-13"
year = date[:4]
month = date[5:7]
day = date[8:]

print(f"Year: {year}, Month: {month}, Day: {day}")

Modifying Strings
--------------------
python
url = "https://example.com"
protocol = url[:5]
domain = url[8:]
print(f"Protocol: {protocol}, Domain: {domain}")
```

```python
Q2. Explain the key features of lists in python.
Answer: Lists in Python are one of the most versatile and widely used data structures. 

Here are some key features that make them so popular:

a) Ordered Collection
---------------------
Lists maintain the order of elements. The first item has an index of 0, the second has an index of 1, and so on
Example:
fruit = ["apple", "banana", "cherry"]
print(fruits[0])  #output: apple.

b) Mutable
------------
Lists can be changed after their creation. You can add, remove, or modify elements.
Example:
fruits = ["apple", "banana", "cherry"]
fruits[1] = "blueberry"
print(fruits)  # Output: ['apple', 'blueberry', 'cherry']

c) Dynamic Size
----------------
Lists can grow or shrink in size as needed, making them flexible for various use cases.
Example:
numbers = [1, 2, 3]
numbers.append(4)
print(numbers)  # Output: [1, 2, 3, 4]
numbers.remove(2)
print(numbers)  # Output: [1, 3, 4]

d) Heterogeneous Elements
---------------------------
Lists can contain elements of different data types, including integers, strings, and even other lists.
Example:
mixed_list = [1, "apple", [2, 3]]
print(mixed_list)  # Output: [1, 'apple', [2, 3]]

e) Slicing
-------------
Lists support slicing, allowing you to access a subset of the list.
Example:
numbers = [1, 2, 3, 4, 5]
print(numbers[1:4])  # Output: [2, 3, 4]

f) Built-in Functions and Methods
----------------------------------
Python provides many built-in functions and methods to work with lists, such as len(), append(), extend(), pop(), remove(), and more.
Example:
fruits = ["apple", "banana", "cherry"]
print(len(fruits))  # Output: 3

fruits.extend(["date", "elderberry"])
print(fruits)  # Output: ['apple', 'banana', 'cherry', 'date', 'elderberry']

fruits.pop()
print(fruits)  # Output: ['apple', 'banana', 'cherry', 'date']

g) List Comprehensions
------------------------
Python supports list comprehensions, a concise way to create lists.
Example:
squares = [x**2 for x in range(5)] print(squares) # Output: [0, 1, 4, 9, 16]
```

```python
Q3. Describe how to access, modify, and delet elements in a list with examples.
Answer: Sure! Working with lists is fundamental in Python, and hereâ€™s how you can access, modify, and delete elements.

a) Accessing Elements
----------------------
You can access elements in a list using their index.
Example:

# List of fruits
fruits = ["apple", "banana", "cherry"]

# Accessing elements by index
first_fruit = fruits[0]
second_fruit = fruits[1]
last_fruit = fruits[-1]

print(first_fruit)  # Output: apple
print(second_fruit) # Output: banana
print(last_fruit)   # Output: cherry

b) Modifying Elements
-----------------------
You can modify elements by assigning new values to the specified index.
Example:

# List of fruits
fruits = ["apple", "banana", "cherry"]

# Modifying the second element
fruits[1] = "blueberry"

print(fruits)  # Output: ['apple', 'blueberry', 'cherry']

=> Adding Elements
You can add elements to a list using append(), insert(), or extend() methods.
Example:

# List of fruits
fruits = ["apple", "banana", "cherry"]

# Adding an element to the end of the list
fruits.append("date")

# Adding an element at a specific position
fruits.insert(1, "blueberry")

# Extending the list with another list
fruits.extend(["elderberry", "fig"])

print(fruits)  # Output: ['apple', 'blueberry', 'banana', 'cherry', 'date', 'elderberry', 'fig']

c) Deleting Elements
----------------------
You can delete elements using del, remove(), or pop() methods.
Example:

# List of fruits
fruits = ["apple", "banana", "cherry", "date"]

# Removing an element by index
del fruits[1]

# Removing an element by value
fruits.remove("cherry")

# Removing the last element
last_fruit = fruits.pop()

print(fruits)    # Output: ['apple']
print(last_fruit)  # Output: date

=> Slicing to Modify and Delete
You can also use slicing to modify or delete multiple elements.
Example:

# List of fruits
fruits = ["apple", "banana", "cherry", "date", "elderberry"]

# Modifying multiple elements
fruits[1:3] = ["blueberry", "cranberry"]

# Deleting multiple elements
del fruits[2:4]

print(fruits)  # Output: ['apple', 'blueberry', 'elderberry']
```

```python
Q4. Compare and contrast tuples and lists with examples.
Answer: Tuples and lists are both used to store collections of items in Python, but they have some key differences.
        Here's a comparison of their main features

Similarities:
---------------
Order: Both tuples and lists are ordered collections, meaning the items have a defined order and can be accessed via indexing.
Heterogeneous: Both can store heterogeneous data, allowing for different data types within the same collection.
Indexing and Slicing: Both support indexing and slicing operations to access elements.

Differences:
---------------
Mutability: Lists are mutable, meaning their elements can be changed after creation,
            while tuples are immutable, so their elements cannot be altered once defined.
Syntax: Lists are defined using square brackets [ ], whereas tuples use parentheses ( ) for their definition.
Size:  The size of a list can be modified by adding or removing elements, making it variable
       while the size of a tuple is fixed and cannot be changed after it is created.
Performance: Lists are generally slower due to their mutable nature, which requires additional memory management
             whereas tuples are faster because they are immutable and require less overhead.
Use Cases: Lists are suitable for dynamic collections where the data needs to change frequently,
           while tuples are ideal for fixed collections where data consistency is important and modifications are not needed

List Example:
--------------
# Creating a list
fruits = ["apple", "banana", "cherry"]
# Modifying an element
fruits[1] = "blueberry"
# Adding an element
fruits.append("date")
# Removing an element
fruits.remove("apple")
print(fruits)  # Output: ['blueberry', 'cherry', 'date']

Tuple Example:
---------------
# Creating a tuple
fruits = ("apple", "banana", "cherry")
# Trying to modify an element will raise an error
try:
    fruits[1] = "blueberry"
except TypeError as e:
    print(e)  # Output: 'tuple' object does not support item assignment
# Accessing elements
print(fruits[1])  # Output: 'banana'
```

```python
Q5. Describe the key features of sets and provides examples of their use.
Answer: Sets in Python are an unordered collection of unique elements. Sets are particularly useful when you need to maintain a collection of 
        unique items and perform operations like union, intersection, and difference efficiently.

Here are some of the key features and examples of their use:

Key Features of Sets:
-----------------------
Unordered: Sets do not maintain the order of elements. The items in a set are stored in an arbitrary order, and their order can change over time.
Unique Elements: Sets automatically remove duplicate elements. Each element in a set is unique.
Mutable: Although the elements themselves must be immutable (like strings, numbers, tuples), sets are mutable, meaning you can add or remove elements.
Unindexed: You cannot access set elements by an index or a key because they are unordered. Instead, you iterate through them.
Dynamic: Sets can grow and shrink as needed, just like lists.
Efficient Membership Testing: Sets provide a highly optimized method for checking if an element is in the set.
    
Examples:

Creating a Set:
----------------
# Creating a set with unique elements
my_set = {1, 2, 3, 4, 4, 5}
print(my_set)  # Output: {1, 2, 3, 4, 5}

Adding Elements:
-----------------
# Adding elements to a set
my_set.add(6)
print(my_set)  # Output: {1, 2, 3, 4, 5, 6}

Removing Elements:
--------------------
# Removing elements from a set
my_set.remove(3)
print(my_set)  # Output: {1, 2, 4, 5, 6}

# Discard method doesn't raise an error if the element is not found
my_set.discard(10)  # No error, even though 10 is not in the set

Membership Testing:
----------------------
# Check if an element is in the set
print(4 in my_set)  # Output: True
print(10 in my_set) # Output: False

Set Operations:
----------------
Union: Combines elements from two sets.

set_a = {1, 2, 3}
set_b = {3, 4, 5}
union_set = set_a | set_b
print(union_set)  # Output: {1, 2, 3, 4, 5}

Intersection: Returns elements that are common to both sets.

intersection_set = set_a & set_b
print(intersection_set)  # Output: {3}
         
Difference: Returns elements that are in one set but not the other.

difference_set = set_a - set_b
print(difference_set)  # Output: {1, 2}

Symmetric Difference: Returns elements that are in either of the sets but not in both.

sym_diff_set = set_a ^ set_b
print(sym_diff_set)  # Output: {1, 2, 4, 5}

Iterating Over a Set:
----------------------
# Iterating over set elements
for element in my_set:
    print(element)
```

```python
Q7. Describe how to add, modify and delete items in a dictionary with examples.
Answer: Dictionaries in Python are a collection of key-value pairs, making them incredibly useful for mapping and organizing data. 

a) Adding Items
To add an item to a dictionary, you simply assign a value to a new key.
---------------------------------------------------------------------------
Example:
# Create an empty dictionary
my_dict = {}
# Add items
my_dict["name"] = "Alice"
my_dict["age"] = 30
my_dict["city"] = "Pune"
print(my_dict)  # Output: {'name': 'Alice', 'age': 30, 'city': 'Pune'}

b) Modifying Items
To modify an item in a dictionary, you reassign a value to an existing key.
----------------------------------------------------------------------------
Example:
# Existing dictionary
my_dict = {"name": "Alice", "age": 30, "city": "Pune"}
# Modify an item
my_dict["age"] = 31
my_dict["city"] = "Mumbai"
print(my_dict)  # Output: {'name': 'Alice', 'age': 31, 'city': 'Mumbai'}

c) Deleting Items
You can delete items from a dictionary using the del keyword or the pop() method.
-----------------------------------------------------------------------------------
Example using del:
# Existing dictionary
my_dict = {"name": "Alice", "age": 31, "city": "Mumbai"}
# Delete an item
del my_dict["age"]
print(my_dict)  # Output: {'name': 'Alice', 'city': 'Mumbai'}
Example using pop():
# Existing dictionary
my_dict = {"name": "Alice", "age": 31, "city": "Mumbai"}
# Delete an item and get its value
age = my_dict.pop("age")
print(my_dict)  # Output: {'name': 'Alice', 'city': 'Mumbai'}
print(age)      # Output: 31

d) Additional Operations
Clear All Items: To remove all items from a dictionary, use the clear() method.
--------------------------------------------------------------------------------
my_dict.clear()
print(my_dict) 
```
